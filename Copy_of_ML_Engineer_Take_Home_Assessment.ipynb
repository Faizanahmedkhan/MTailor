# MODEL IMPLEMENTATION:

```python
import os
import torch
import torch.nn as nn
import numpy as np

from torchvision.models import resnet18, ResNet18_Weights
```

```python
class Model(nn.Module):

  def __init__(self):
    super(Model, self).__init__()
    pretrained_model = resnet18(weights=ResNet18_Weights.DEFAULT)
    self.backbone = nn.Sequential(*list(pretrained_model.children())[:-2])
    self.maxpool = nn.AdaptiveMaxPool2d((1, 1))  # output size should be (1,1)
    self.fc = nn.Linear(512, 512)  # input size should match the output of backbone

   ### Initialize the required Layers
    self.have_object = nn.Linear(512, 2)
    self.cat_or_dog = nn.Linear(512, 3)
    self.specie = nn.Linear(512, 10)
    self.bbox = nn.Linear(512, 4)


  def forward(self, input):
      out_backbone = self.backbone(input)
      ### Write Forward Calls for the Model
      out_backbone = self.maxpool(out_backbone)
      out_backbone = out_backbone.view(out_backbone.size(0), -1)  # Flatten the output
      out_backbone = self.fc(out_backbone)
      return {
          "bbox": self.bbox(out_backbone),
          "have_object": self.have_object(out_backbone),
          "cat_or_dog": self.cat_or_dog(out_backbone),
          "specie": self.specie(out_backbone)
      }


```

# CUSTOM DATALOADER IMPLEMENTATION

```python
train_list = np.load('/content/assessment_dataset/train_list.npy', allow_pickle=True).tolist()
val_list = np.load('/content/assessment_dataset/val_list.npy', allow_pickle=True).tolist()
```

```python
from bs4 import BeautifulSoup

def read_xml_file(path):
    with open(path, 'r') as f:
        data = f.read()
    bs_data = BeautifulSoup(data, 'xml')
    return {
        "cat_or_dog": bs_data.find("name").text,
        "xmin": int(bs_data.find("xmin").text),
        "ymin": int(bs_data.find("ymin").text),
        "xmax": int(bs_data.find("xmax").text),
        "ymax": int(bs_data.find("ymax").text),
        "specie": "_".join(path.split(os.sep)[-1].split("_")[:-1])
    }
```

```python
from PIL import Image

import torchvision.transforms as transforms

import pandas as pd


class CustomDataset():

  def __init__(self, dataset_path, images_list, train=False):

    self.preprocess = transforms.Compose([
    transforms.Resize((224, 224)),  
    ])
    self.data = []

    image_folder_path = os.path.join(dataset_path, "images")
    label_folder_path = os.path.join(dataset_path, "labels")

    for path in os.listdir(image_folder_path):
        name = path.split(os.sep)[-1].split(".")[0]
        if name in images_list:
          try:
            xml_path = os.path.join(label_folder_path, name+".xml")
            xml_data = read_xml_file(xml_path)
            xml_data['have_object']=1

          except:
            xml_data={'cat_or_dog': 'Unknown', 'xmin': 0, 'ymin': 0, 'xmax': 0, 'ymax': 0, 'specie': 'Unknown'}
            xml_data['have_object']=0


          if xml_data['cat_or_dog']=='cat':
            xml_data['cat_or_dog']=torch.tensor(1)
          elif xml_data['cat_or_dog']=='dog':
            xml_data['cat_or_dog']=torch.tensor(2)
          elif xml_data['cat_or_dog']=='Unknown':
            xml_data['cat_or_dog']=torch.tensor(0)

          dictionary={
            "Unknown":torch.tensor(0),
            "Abyssinian":torch.tensor(1),
            "Birman":torch.tensor(2),
            "Persian":torch.tensor(3),
            "american_bulldog":torch.tensor(4),
            "american_pit_bull_terrier":torch.tensor(5),
            "basset_hound":torch.tensor(6),
            "beagle":torch.tensor(7),
            "chihuahua":torch.tensor(8),
            "pomeranian":torch.tensor(9)
          }

          xml_data['specie']=dictionary[xml_data['specie']]
          xml_data['bbox'] = torch.tensor([xml_data['xmin'], xml_data['ymin'], xml_data['xmax'], xml_data['ymax']])
          xml_data['image_path'] = os.path.join(dataset_path, "images", path)

          self.data.append(xml_data)

    df= pd.DataFrame(self.data)
    self.temp_have = nn.functional.one_hot(torch.tensor(df['have_object']))
    self.temp_specie = nn.functional.one_hot(torch.tensor(df['specie']))
    self.temp_cat_dog = nn.functional.one_hot(torch.tensor(df['cat_or_dog']))

    for i in range(len(self.data)):
      self.data[i]['have_object'] = self.temp_have[i]
      self.data[i]['specie'] = self.temp_specie[i]
      self.data[i]['cat_or_dog'] = self.temp_cat_dog[i]


  def __len__(self):
    return len(self.data)

  def __getitem__(self, index):
    new_width = 224
    new_height = 224

    img_path = self.data[index]['image_path']

    image = Image.open(img_path)     
    original_width , original_height = image.size

    width_scale = new_width / original_width
    height_scale = new_height / original_height


    self.data[index]['xmin'] = self.data[index]['xmin'] * width_scale
    self.data[index]['ymin'] = self.data[index]['ymin']  * height_scale
    self.data[index]['xmax'] = self.data[index]['xmax']  * width_scale
    self.data[index]['ymax'] = self.data[index]['ymax']  * height_scale


    if self.preprocess:
      image = self.preprocess(image)

    label = {
        'cat_or_dog': self.data[index]['cat_or_dog'],
        'xmin': self.data[index]['xmin'], 
        'ymin': self.data[index]['ymin'], 
        'xmax': self.data[index]['xmax'], 
        'ymax': self.data[index]['ymax'], 
        'specie': self.data[index]['specie'],
        'have_object': self.data[index]['have_object'],
        'bbox': self.data[index]['bbox'] 
    }

    image = self.preprocess(image)
    return transforms.ToTensor()(image), label
```

```python

def test(model, val_list):

  def post_process_object(x):
    return torch.where(x > 0.5, 1.0, 0.0).squeeze(1)

  def post_process_cat_or_dog(x):
    return torch.argmax(x,dim=1)

  def post_process_specie(x):
    return torch.argmax(x, dim=1)


  def __tl(x):
    return x.tolist()

  def __tn(x):
    return x.detach().cpu().numpy()

  def __tnl(x):
    return (x.detach().cpu().numpy()).tolist()

  metric_object = torchmetrics.F1Score(task="binary")
  metric_cat_or_dog = torchmetrics.F1Score(task="multiclass", num_classes=3)
  metric_specie = torchmetrics.F1Score(task="multiclass", num_classes=10)

  output_list = {
      "object": [],
      "cat_or_dog": [],
      "specie": [],
      "xmin": [],
      "ymin": [],
      "xmax": [],
      "ymax": [],
  }
  labels_list = {
      "object": [],
      "cat_or_dog": [],
      "specie": [],
      "xmin": [],
      "ymin": [],
      "xmax": [],
      "ymax": [],
  }


  validation_dataset = CustomDataset("/content/assessment_dataset",val_list)
  val_loader = torch.utils.data.DataLoader(validation_dataset, batch_size=32, shuffle=True)


  for i, data in enumerate(val_loader):
    inputs, labels = data
    if torch.cuda.is_available():
      inputs = inputs.cuda()
      # for key, value in labels.items():
      #     labels[key] = labels[key].cuda()

    # Make predictions for this batch
    outputs = model(inputs)
    labels["width"] = labels["xmax"]-labels["xmin"]
    labels["height"] = labels["ymax"]-labels["ymin"]
    is_object = __tnl(labels["have_object"])
    width = __tn(labels["width"])
    height = __tn(labels["height"])
    output_list["object"].extend(__tnl(post_process_object(outputs["have_object"])))
    labels_list["object"].extend(__tnl(labels["have_object"]))



    output_list["cat_or_dog"].extend(
      __tnl(post_process_cat_or_dog(outputs["cat_or_dog"]))
    )
    labels_list["cat_or_dog"].extend(
      __tnl(post_process_cat_or_dog(labels["cat_or_dog"]))
    )
    output_list["specie"].extend(
      __tnl(post_process_specie(outputs["specie"]))
    )
    labels_list["specie"].extend(__tnl(post_process_specie(labels["specie"])))
    output_list["xmin"].extend(
      __tl(__tn((outputs["bbox"][:, 0]))*width)
    )
    labels_list["xmin"].extend(__tl(__tn(labels["xmin"])*width))
    output_list["ymin"].extend(
        __tl(__tn((outputs["bbox"][:, 1]))*height)
    )
    labels_list["ymin"].extend(
        __tl(__tn(labels["ymin"])*height)
    )
    output_list["xmax"].extend(
        __tl(__tn((outputs["bbox"][:, 2]))*width)
    )
    labels_list["xmax"].extend(__tl(__tn(labels["xmax"])*width))
    output_list["ymax"].extend(__tl(__tn((outputs["bbox"][:, 3]))*height))
    labels_list["ymax"].extend(__tl(__tn(labels["ymax"])*height))


  score_object = metric_object(torch.tensor(output_list["object"]), torch.tensor(labels_list["object"]))
  score_cat_or_dog = metric_cat_or_dog(torch.tensor(output_list["cat_or_dog"]), torch.tensor(labels_list["cat_or_dog"]))
  score_specie = metric_specie(torch.tensor(output_list["specie"]), torch.tensor(labels_list["specie"]))
  
  box1=outputs['bbox'].to('cuda')
  box2=labels['bbox'].to('cuda')
  score_bbox = torchvision.ops.distance_box_iou_loss(box1, box2,reduction='mean')
  print(score_bbox) 

  

  return score_object, score_cat_or_dog, score_specie, score_bbox
```

```python
# import our library
import torchmetrics

def train(epochs, model_weights):
  model = Model()

  if model_weights:
    model.load_state_dict(model_weights)
    # Initialize Model and Optimizer

  optimizer = torch.optim.Adam(model.parameters(),lr = 0.0001)

  # Initialize Loss Functions
  have_object_loss = nn.CrossEntropyLoss()
  specie_loss = nn.CrossEntropyLoss()
  cat_or_dog_loss = nn.CrossEntropyLoss()

  training_dataset = CustomDataset("/content/assessment_dataset", images_list=train_list)
  training_loader = torch.utils.data.DataLoader(training_dataset, batch_size=32, shuffle=True)

  if torch.cuda.is_available():
    model = model.cuda()

  def train_one_epoch(epoch_index, tb_writer):
      running_loss = 0.
      last_loss = 0.

      # Here, we use enumerate(training_loader) instead of
      # iter(training_loader) so that we can track the batch
      # index and do some intra-epoch reporting
      for i, data in enumerate(training_loader):
          # Every data instance is an input + label pair
          inputs, labels = data
          #print(data)
          # Make predictions for this batch
          outputs = model(inputs.to("cuda"))

          # Compute the loss and its gradients
          loss_have_object = have_object_loss(outputs["have_object"], labels['have_object'].type(torch.float32).to('cuda'))
          loss_specie = specie_loss(outputs["specie"], labels['specie'].type(torch.float32).to('cuda'))
          loss_cat_or_dog = cat_or_dog_loss(outputs["cat_or_dog"], labels['cat_or_dog'].type(torch.float32).to('cuda'))
          boxes1 = outputs['bbox'].to("cuda")
          boxes2 = labels['bbox'].to("cuda")
         
          loss_bbox = torchvision.ops.distance_box_iou_loss(boxes1, boxes2, reduction= 'mean') #bbox_loss(outputs["bbox"], labels["bbox"])
          # Above or Below
          #loss_xmin = xmin_loss(outputs["xmin"], labels['xmin'].type(torch.float32).to('cuda'))
          #loss_ymin = ymin_loss(outputs["ymin"], labels['ymin'].type(torch.float32).to('cuda'))
          #loss_xmax = xmax_loss(outputs["xmax"], labels['xmax'].type(torch.float32).to('cuda'))
          #loss_ymax = ymax_loss(outputs["ymax"], labels['ymax'].type(torch.float32).to('cuda'))

          #torchvision.ops.distance_box_iou_loss(boxes1, boxes2: Tensor, reduction: str = 'none', eps: float = 1e-07)
          #print(loss_bbox)
          loss = loss_have_object + loss_specie.type(torch.float32) + loss_cat_or_dog.type(torch.float32) + loss_bbox # Consolidate all individual losses
          #print("loss_bbox ",loss_bbox.sum(),"loss_cat_or_dog ",loss_cat_or_dog,"loss_have_object ",loss_have_object,"loss_specie ",loss_specie)
          loss.backward()

          optimizer.step()

          optimizer.zero_grad()

          # Gather data and report
          running_loss += loss.item()
          if i % 10 == 0:
              last_loss = running_loss / 10 # loss per batch
              running_loss = 0.
      return last_loss
  dataset_path = '/content/assessment_dataset'

  for i in range(epochs):

    epoch_loss = train_one_epoch(i, None)
    print(f' Epoch {i} Loss : {epoch_loss}')

    torch.save( model.state_dict(),"model.pth")
    metrics = test(model,val_list)
    print(metrics)
```

```python
import torch
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt

model = Model()  
model.load_state_dict(torch.load('model.pth'))

# Loading the image
image = Image.open('/content/assessment_dataset/images/american_pit_bull_terrier_129.jpg')

# Preprocessing the image
preprocess = transforms.Resize([224, 224])
image = preprocess(image)
image = transforms.ToTensor()(image)
image = image.unsqueeze(0)

# Performing inference
out = model(image)
obj = torch.argmax(out['have_object'])
cord = torch.argmax(out['cat_or_dog'])
spec = torch.argmax(out['specie'])
bbox = out['bbox']

# Defining class labels
class_labels = {
    0: "Unknown",
    1: "Cat",
    2: "Dog",
}

# Defining species labels
species_labels = {
    0: "Unknown",
    1: "Abyssinian",
    2: "Birman",
    3: "Persian",
    4: "american_bulldog",
    5: "american_pit_bull_terrier",
    6: "basset_hound",
    7: "beagle",
    8: "chihuahua",
    9: "pomeranian",
}

# Converting predictions to labels
obj_label = "Has Object" if obj.item() == 1 else "No Object Found"
cord_label = class_labels.get(cord.item(), "Unknown")
spec_label = species_labels.get(spec.item(), "Unknown")

# Displaying the image
plt.imshow(image.squeeze(0).permute(1, 2, 0))
plt.axis('off')

# Drawing the bounding box
if obj.item() == 1:
    bbox = bbox.squeeze().tolist()  # Convert to list
    plt.gca().add_patch(plt.Rectangle((bbox[0], bbox[1]), bbox[2] - bbox[0], bbox[3] - bbox[1], 
                                      edgecolor='r', facecolor='none'))

# Showing the image with bounding box
plt.show()

# Print the predictions
print("Object:", obj_label)
print("Category:", cord_label)
print("Species:", spec_label)
```
